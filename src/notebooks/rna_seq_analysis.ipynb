```python
res_space = res_space[res_space.baseMean >= 10]
```

```python
res_space
```

```python
sigs = res_space[(res_space.padj < 0.05 ) & (abs(res_space.log2FoldChange) > 0.5) & (res_space.baseMean > 20)]
```


1. `res_space.padj < 0.05`: Select rows where the adjusted p-value (`padj`) is less than 0.05. The adjusted p-value is a statistical measure that accounts for multiple comparisons. In the context of gene expression, an adjusted p-value below 0.05 typically indicates that the result is statistically significant, and the likelihood that the observed data occurred by chance is low.

2. `abs(res_space.log2FoldChange) > 0.5`: Select rows where the absolute value of the log2 fold change (`log2FoldChange`) is greater than 0.5. The log2 fold change is a measure of change in expression levels between two conditions. A value greater than 0.5 suggests that the gene expression has changed by at least \( 2^{0.5} \approx 1.41 \) times, meaning the gene is either upregulated or downregulated by at least that factor.

3. `res_space.baseMean > 20`: Select rows where the base mean expression (`baseMean`) is greater than 20. This condition filters for genes that have a sufficiently high average expression level across all samples; it's used to exclude genes with low expression levels which might not be biologically relevant or could lead to unreliable statistical inference due to high variability.
